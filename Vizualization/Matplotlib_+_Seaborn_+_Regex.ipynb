```
!wget "https://s3.amazonaws.com/thinkific/file_uploads/118220/attachments/81e/d1b/6d0/Apply_function.zip"
!unzip Apply_function.zip
data = pd.read_csv('Apply function/bigmart_data.csv')
```

```
data = data.dropna(how = 'any')
data.head()
```

## Line Chart

```
price_by_item = data.groupby(by = ['Item_Type']).Item_MRP.mean()[0:10]
price_by_item
```

```
x = price_by_item.index.tolist()
y = price_by_item.values.tolist()
plt.figure(figsize=(20, 5))
plt.xlabel=('Item Type')
plt.ylabel('Mean price')
plt.plot(x,y)
```

```
price_by_size = data.groupby('Outlet_Size').Item_Outlet_Sales.mean()
x = price_by_size.index.tolist()
y = price_by_size.values.tolist()
plt.figure(figsize=(10, 5))
plt.xlabel=('Item Type')
plt.ylabel('Mean price')
plt.bar(x,y)
```

# Seaborn

```
#Build on top of matplotlib
import seaborn as sns
sns.set(style = 'darkgrid')
data['Item_Visibility'] = data['Item_Visibility'] *100
data.head()
```

```
sns.lineplot(x = 'Item_Weight', y = 'Item_MRP', data=data[0:50])
```

```
sns.barplot(x = 'Item_Fat_Content', y = 'Item_MRP', data=data[0:50])
```

```
sns.distplot(data['Item_MRP'])
```

```
sns.boxplot(data['Item_Outlet_Sales'])
```

```
sns.boxplot(data['Item_MRP'])
```

```
sns.violinplot(data['Item_Outlet_Sales'])
```

```
sns.relplot(y='Item_Outlet_Sales', x='Item_MRP', kind='scatter', data=data[0:200])
```

```
sns.relplot(y='Item_Outlet_Sales', x='Item_MRP', hue = 'Item_Type', kind='scatter', data=data[0:200])
```

```
plt.figure(figsize = (10, 5))
sns.relplot(y='Item_Outlet_Sales', x='Item_MRP', size = 'Item_Visibility', hue = 'Item_Visibility', kind='scatter', data=data[0:200])
```

## Advanced sns

```
sns.catplot(x='Outlet_Size', y ='Item_Outlet_Sales',kind = 'strip', data = data[0:500])
```

```
sns.catplot(x='Outlet_Size', y ='Item_Outlet_Sales',kind = 'swarm', data = data[0:500])
```

```
sns.catplot(x='Outlet_Size', y ='Item_Outlet_Sales',kind = 'box', data = data[0:500])
```

```
sns.catplot(x='Outlet_Size', y ='Item_Outlet_Sales',kind = 'violin', data = data[0:500])
```

```
sns.pairplot(data = data[0:350], hue = 'Outlet_Size', height = 2.5)
```
